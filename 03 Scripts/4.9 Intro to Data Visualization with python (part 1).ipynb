```python
# Importing customers dataframe
df_customers=pd.read_csv(('../02 Data/Original Data/customers.csv'),index_col=False)
```

# 03 Data Wringling

```python
df_customers.head()
```

```python
df_customers.describe()
```

```python
# Change types for df_customers data set
df_customers['user_id']=df_customers['user_id'].astype('int32')
df_customers['Age']=df_customers['Age'].astype('int8')
df_customers['n_dependants'] = df_customers['n_dependants'].astype('int8')
df_customers['income']=df_customers['income'].astype('int32')
df_customers['Gender']=df_customers['Gender'].astype('category')
df_customers['STATE']=df_customers['STATE'].astype('category')
df_customers['fam_status']=df_customers['fam_status'].astype('category')
```

```python
df_customers.info()
```

```python
# Ranaming n_dependants column
df_customers.rename(columns={'First Name':'first_name','Surnam':'last_name','Gender':'gender','STATE':'state','Age':'age','n_dependants':'number_of_dependants','fam_status':'family_status'}, inplace=True)
```

# 04 Data Cosistency Checks

```python
#Check for mixed data type
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
#change the mixed data type as approprate:
df_customers['first_name']=df_customers['first_name'].astype('str')
```

```python
# Finding Missing data
df_customers.isnull().sum()
```

```python
# Finding Duplicates
df_customers_dups=df_customers[df_customers.duplicated()]
```
