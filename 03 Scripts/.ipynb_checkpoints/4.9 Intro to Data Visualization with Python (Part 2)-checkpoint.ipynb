# 01 Importing libraries

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# 02 Importing dataframe

```python
# Importing ords_prods_cust dataframe
ords_prods_cust=pd.read_pickle('../02 Data/Prepared Data/orders_products_customers.pkl')
```

# 03 Creating Charts

## Histogram

```python
plt.ticklabel_format(style='plain')
hist=ords_prods_cust['prices'].plot.hist(bins = 25)
plt.title('Frequency for Prices Ranges (25 bins)')
```

```python
plt.ticklabel_format(style='plain')
hist2=ords_prods_cust['prices'].plot.hist(bins = 70)
plt.title('Frequency for Prices Ranges (70 bins)')
```

```python
# Exporting Histogram graph
hist.figure.savefig ('../04 Analysis/Visualizations/Histogram_Prices.png')
hist2.figure.savefig ('../04 Analysis/Visualizations/Histogram_Prices_2.png')
```

## Bar

```python
# Create Bar Chart without data arranged from largest to smallest
plt.ticklabel_format(style='plain')
bar=ords_prods_cust['orders_day_of_week'].value_counts().sort_index().plot.bar()
plt.title('Number of Orders by Day of the Week')
```

```python
# Exporting Bar graph
bar.figure.savefig('../04 Analysis/Visualizations/bar_orders_dow.png')
```

```python
#Scatterplot of the prices column
scat = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_cust)
plt.title('Prices Distribution points')
```

```python
#Export Scatterplot
scat.figure.savefig('../04 Analysis/Visualizations/scat_prices.png')
```

# 04 Sampling Data

```python
# Create a list holding True/False values to the test np.random.rant()<=0.7
dev=np.random.rand(len(ords_prods_cust))<=0.7
```

```python
# Store 70% of the sample in the dataframe big
big=ords_prods_cust[dev]
```

```python
# Store 30% of the sample in the dataframe small
small=ords_prods_cust[~dev]
```

```python
len(ords_prods_cust)
```

```python
len(big)+len(small)
```

```python
# Create a small sample
df_1 = small[['orders_day_of_week','prices']]
```

```python
# Create a line chart
line_1= sns.lineplot(data = df_1, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting line graph
line_1.figure.savefig('../04 Analysis/Visualizations/line_orders_day_per_prices.png')
```

```python
# Create Histogram Chart for number of orders per hour
plt.ticklabel_format(style='plain')
hist_3=ords_prods_cust['order_hour_of_day'].plot.hist(bins = 24)
plt.title('Number of Orders by Hour of the Day')
```

Q3. The above Histogram shows that the busiest time in the day is between 10 and 15 hrs. which we can consider increasing workers number during that time.

```python
# Exporting Histogram graph
hist_3.figure.savefig('../04 Analysis/Visualizations/Histogram_order_hour_of_day.png')
```

```python
# Create Bar Chart for loyalty flag column
plt.ticklabel_format(style='plain')
bar_2=ords_prods_cust['loyalty_flag'].value_counts().sort_index().plot.bar()
plt.title('Number of Orders by Customer Loyalty')
```

```python
# Exporting Bar graph
bar_2.figure.savefig(('../04 Analysis/Visualizations/loyalty_flag_bar.png'),bbox_inches='tight')
```
