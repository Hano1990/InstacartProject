# 1 Importing libraries

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# Importing dataframe

```python
#Importing ords_prods_all dataframe
ords_prods_all=pd.read_pickle('../02 Data/Prepared Data/orders_products_all.pkl')
```

```python
# Create Bar Chart for price range loc column
plt.ticklabel_format(style='plain')
price_range=ords_prods_all['price_range_loc'].value_counts().sort_index().plot.bar()
plt.title('Number of Orders per Price Range')
```

```python
# Exporting Bar graph
price_range.figure.savefig(('../04 Analysis/Visualizations/price_range.png'),bbox_inches='tight')
```

```python
# Check customer profile loyalty with department

loyalty_department = pd.crosstab(ords_prods_all['department'],ords_prods_all['loyalty_flag'],dropna=False)
```

```python
loyalty_department = loyalty_department.sort_values(by=['Regular customer'], ascending=True)
```

```python
# Create department orders by Loylity flag bar chart
Loyalty_department_bar = loyalty_department.plot.bar(stacked = True)
plt.title('Number of Orders by Customer Loyalty and Department')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# Exporting Bar graph
Loyalty_department_bar.figure.savefig(('../04 Analysis/Visualizations/Loyalty_department_bar'),bbox_inches='tight')
```

```python
family_status_bar = ords_prods_all['family_status'].value_counts().plot.bar()
plt.title('Number of Orders by Family Status')
plt.show()
```

```python
# Exporting Bar graph
family_status_bar.figure.savefig(('../04 Analysis/Visualizations/family_status_bar'),bbox_inches='tight')
```

```python
# Check Parental_profile with income Category

Parental_income = pd.crosstab(ords_prods_all['income_category'],ords_prods_all['Parental_profile'],dropna=False)
```

```python
# Create Parental_profile with income Category bar chart
Parental_income_bar = Parental_income.plot.bar(stacked = True)
plt.title('Number of Orders by Parental Profile and Income Category')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# Exporting Bar graph
Parental_income_bar.figure.savefig(('../04 Analysis/Visualizations/Parental_income_bar'),bbox_inches='tight')
```

```python
# Check family status with region

family_region = pd.crosstab(ords_prods_all['family_status'],ords_prods_all['region'],dropna=False)
```

```python
# Create family status with region bar chart
family_region_bar = family_region.plot.bar(stacked = True)
plt.title('Family Status by Region')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# Exporting Bar graph
family_region_bar.figure.savefig(('../04 Analysis/Visualizations/family_region_bar'),bbox_inches='tight')
```
